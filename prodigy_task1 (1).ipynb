```python

def average(data):
    avg_values = {}
    years = data.loc[:,'1970':'2021']
    
    for year in years:
        avg_values[year] = data[year].mean()
    
    return avg_values

avg_population = average(data)
print(avg_population)
```

```python

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(avg_population.keys(), avg_population.values(), color='blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Average Population')
plt.xticks(rotation=90)
plt.show()

```

```python

countries = ['AFG', 'CHN', 'IND', 'PAK', 'MEX']  # Example countries

# Choose the indicator
indicator_name = 'Population, total'  # Population indicator

# Extract years and corresponding population values for the selected countries
years = range(1970, 2023)
country_data = data[data['Indicator Name'] == indicator_name]
country_data = country_data[country_data['Country Code'].isin(countries)]
country_data = country_data.set_index('Country Code')
country_data = country_data.loc[countries, '1970':]

# Plot the population trends for the selected countries
plt.figure(figsize=(10, 6))  #size of the graph
for country in countries:
    plt.plot(years, country_data.loc[country], label=country)

# Set labels and title
plt.xlabel('Year')
plt.ylabel(indicator_name)
plt.title(f'{indicator_name} Trends for Selected Countries')
plt.legend()
plt.grid(True)
plt.xticks(range(1970, 2023, 5))  # Show ticks every 5 years for better readability
plt.tight_layout()

plt.show()

     
```
