```python
import pandas as pd
import matplotlib.pyplot as plt

# Specify the file path (using forward slashes or double backslashes)
file_path = r'D:\My Data\Downloads\09cad793-22f9-46b3-9f77-195e049a526d_Data.csv'

# Read the CSV file, specifying the encoding if needed
data = pd.read_csv(file_path, encoding='utf-8')

print(data.head())
```

```python
data1=data.dropna()
print(data1.head())
```

```python

def average(data1):
    avg_values = {}
    years = ['2011 [YR2011]', '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', 
             '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', 
             '2019 [YR2019]', '2020 [YR2020]']
    
    for year in years:
        avg_values[year] = data1[year].mean()
    
    return avg_values

avg_CO2_emissions = average(data1)

print(avg_CO2_emissions)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(avg_CO2_emissions.keys(), avg_CO2_emissions.values(), color='red')
plt.xlabel('Year')
plt.ylabel('Average CO2 Emission')
plt.title('Average CO2 Emission for Each Year')
plt.xticks(rotation=45)
plt.show()
```

```python
def minimum(data1):
    min_values = {}
    years = ['2011 [YR2011]', '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', 
             '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', 
             '2019 [YR2019]', '2020 [YR2020]']
    
    for year in years:
        min_values[year] = data1[year].min()
    
    return min_values

minimum_CO2_emissions = minimum(data1)
print(minimum_CO2_emissions)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(minimum_CO2_emissions.keys(), minimum_CO2_emissions.values(), color='green')
plt.xlabel('Year')
plt.ylabel('Minimum CO2 Emission')
plt.title('Minimum CO2 Emission for Each Year')
plt.xticks(rotation=45)
plt.show()
```
